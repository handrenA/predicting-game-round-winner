## preprocessing 

```python
# target variable 
target = 'round_winner'

y = df[target]


X = df[['time_left', 'ct_score', 't_score', 'map', 'bomb_planted', 'ct_health',
       't_health', 'ct_armor', 't_armor', 'ct_money', 't_money', 't_helmets', 
        'ct_defuse_kits', 'ct_players_alive', 't_players_alive' ]]
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer

# numerical columns
num_features = X.select_dtypes('number').columns
num_transformer = make_pipeline(SimpleImputer(strategy='most_frequent'), StandardScaler())

cat_features = ['map', 'bomb_planted']
cat_transformer = make_pipeline(
            SimpleImputer(strategy='most_frequent'), 
            OneHotEncoder(handle_unknown='error', drop='if_binary') 
            )

preprocessor = make_column_transformer(
            (num_transformer, num_features), 
            (cat_transformer, cat_features) 
            )
```

## Model selection

```python
# finding the model with highest accuracy score

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.model_selection import GridSearchCV



classifiers = [RandomForestClassifier(),
               SVC(),
              LogisticRegression(),
              SGDClassifier()]

for classes in classifiers:
    # find the model with the highest ROC_AUC score
    pipe = make_pipeline(preprocessor, classes)
    grid = GridSearchCV(estimator = pipe, 
                        param_grid = {}, 
                        cv = 5,
                        scoring = 'roc_auc',
                        error_score='raise')
    grid.fit(X_train, y_train)
    score = grid.best_score_
    print(f'Classifier {classes} with score {score}')
    
```

## Proceeding with Random Forest Classifier 

### GridSearch or RandomizedSearch? 

```python
# hypertuning parameters using Grid Search CV and Randomized Search CV
# we will proceed with the best model 

# Grid Search 

param_grid_RFC = { 'randomforestclassifier__n_estimators' : [50, 100,200], 
             'randomforestclassifier__min_samples_leaf' : [1,2,3],
             'randomforestclassifier__max_depth' : [10, None]}


RFC = RandomForestClassifier()
pipe = make_pipeline(preprocessor, RFC)

grid_RFC = GridSearchCV(estimator = pipe, 
                        param_grid = param_grid_RFC, 
                        cv = 5,
                        scoring = 'roc_auc',
                        error_score='raise')

grid_RFC.fit(X_train, y_train)

print(f'Best score: {grid_RFC.best_score_}\n Best params: {grid_RFC.best_params_}')
```

```python
# hypertuning parameters using Grid Search CV and Randomized Search CV
# we will proceed with the best model 

# Grid Search 



RFC = RandomForestClassifier(n_estimator = 200, min_samples_leaf=1, max_depth=None)
pipe = make_pipeline(preprocessor, RFC)

grid_RFC = GridSearchCV(estimator = pipe, 
                        param_grid = param_grid_RFC, 
                        cv = 5,
                        scoring = 'roc_auc',
                        error_score='raise')

grid_RFC.fit(X_train, y_train)

print(f'Best score: {grid_RFC.best_score_}\n Best params: {grid_RFC.best_params_}')
```

```python
# Randomized Search
from sklearn.model_selection import RandomizedSearchCV


RFC = RandomForestClassifier()
pipe = make_pipeline(preprocessor, RFC)


random_RFC = RandomizedSearchCV(estimator = pipe, 
                        param_distributions = param_grid_RFC, 
                        cv = 5,
                        scoring = 'roc_auc',
                        error_score='raise')

random_RFC.fit(X_train, y_train)

print(f'Best score: {random_RFC.best_score_}\n Best params: {random_RFC.best_params_}')
```

```python
# We can conclude that there's almost no difference between Grid and Randomized, and therefore
# we proceed with the Grid Search Crossvalid score on the training data of 0.9263
```

## Final evaulation on the test data with ROC-AUC plot 

```python
# Actual score on the test data

trained_models = {
    'Randomized Search' : grid_RFC,
    'Grid Search' : random_RFC,
}

for name, model in trained_models.items():
    score = model.score(X_test, y_test)
    params = model.best_params_
    
    print(f'Score for {name}: {score}\n')
```

```python
from sklearn.metrics import RocCurveDisplay
plt.figure(figsize=(10,10))

# Roc Curve for GRID SEARCH 
RocCurveDisplay.from_estimator(grid_RFC, X_test, y_test)

# Roc Curve for Randomized Search 
RocCurveDisplay.from_estimator(random_RFC, X_test, y_test)
plt.show()
```
